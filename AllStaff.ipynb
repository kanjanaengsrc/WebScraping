```
import urllib.parse
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```
def readHtml(url):
  headers = {'User-Agent':'Scraping-Bot'}
  response = requests.get(url,headers=headers)
  if(response.status_code != 200):
    return None
  else:
    response.encoding = 'utf-8'
  return response.text.replace('\n','')
```

```
def getFullname(soup):
  fullname = []
  marker = soup.find('div',text='ชื่อ-สกุล')
  if(marker == None):
    return None
  else:
    rows = marker.parent.find_all_next('p')
    for row in rows:
      if(row.find('a') != None):
        fullname.append(row.a.get_text())
    return fullname[0:-2]
```

```
def getPosition(soup):
  position = []
  marker = soup.find('p',string='ตำแหน่ง')
  for val in marker.find_all_next('p'):
    position.append(val.get_text())
  return position[0:-3]
```

```
def gotoDept(faculty,soup):
  #Inside each value of dictionary
  for val,url in faculty.items():
    fullname = []
    position = []
    member = {}
    data = readHtml(url)
    soup = BeautifulSoup(data,'html.parser')
    print('\n-------',val,'---------\n')
    fullname = getFullname(soup)
    if(fullname != None):
      position = getPosition(soup)
      member.update(zip(fullname,position))
      for name,pos in member.items():
        print(name,'<--->',pos)
      sleep(1)
    else:
      print('**Scraping bot do not work propoerly**')
```

```
url = "http://www.eng.src.ku.ac.th/eng_cpe/"
data = readHtml(url)
soup = BeautifulSoup(data,'html.parser')
faculty = {}
```

```
marker = soup.find('h3',text='บุคลากร')
rows = marker.find_all_next('li')
for dept in rows:
  faculty.update({dept.text:dept.a['href']})
gotoDept(faculty,soup)
```
